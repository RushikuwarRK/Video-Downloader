```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Bidirectional, TimeDistributed, Dropout
from sklearn.model_selection import train_test_split

# Load dataset
df = pd.read_csv('your_dataset1.csv')  # Replace with your actual CSV file path

# Ensure your CSV has columns like 'input' and 'response'
df['input'] = df['input'].fillna('').astype(str)
df['response'] = df['response'].fillna('').astype(str)

inputs = df['input'].tolist()
responses = df['response'].tolist()

# Tokenizing and padding the text data
vocab_size = 5000  # Vocabulary size
oov_token = "<OOV>"
max_length = 50  # Maximum length of sequences

tokenizer = Tokenizer(num_words=vocab_size, oov_token=oov_token)
tokenizer.fit_on_texts(inputs + responses)
input_sequences = tokenizer.texts_to_sequences(inputs)
response_sequences = tokenizer.texts_to_sequences(responses)

input_padded = pad_sequences(input_sequences, maxlen=max_length, padding='post', truncating='post')
response_padded = pad_sequences(response_sequences, maxlen=max_length, padding='post', truncating='post')

# Define the model
embedding_dim = 64
model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_length),
    Bidirectional(LSTM(128, return_sequences=True)),
    Dropout(0.2),
    Bidirectional(LSTM(64, return_sequences=True)),
    TimeDistributed(Dense(vocab_size, activation='softmax'))
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Prepare targets for training
y_train = np.expand_dims(response_padded, axis=-1)

# Split data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(input_padded, y_train, test_size=0.2, random_state=42)

# Train the model
model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_val, y_val))

# Function to generate a response from the model
def generate_response(input_text):
    input_sequence = tokenizer.texts_to_sequences([input_text])
    input_padded = pad_sequences(input_sequence, maxlen=max_length, padding='post')
    predicted_sequence = model.predict(input_padded, verbose=0)

    # Convert predicted indices back to text
    predicted_indices = np.argmax(predicted_sequence, axis=2)[0]
    response_text = tokenizer.sequences_to_texts([predicted_indices])[0]

    return response_text

# Interactive chat loop
def chat():
    print("Chatbot is ready! Type 'exit' to stop.")
    while True:
        user_input = input("You: ")
        if user_input.lower() == 'exit':
            print("Goodbye!")
            break
        response = generate_response(user_input)
        print(f"Chatbot: {response}")

# Start the chat
chat()
```
